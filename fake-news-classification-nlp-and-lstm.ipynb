```python
X=df.drop('label',axis=1)
y=df['label']
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
corpus=[]
for i in range(0,8000):
    ele=re.sub('[^a-zA-Z]',' ',message.title.iloc[i])
    ele=ele.lower()
    ele=ele.split()
    ele=[ps.stem(word) for word in ele if not word in stopwords.words('english')]
    ele=' '.join(ele)
    corpus.append(ele)
```

```python
corpus
```

```python
corpus[1]
```

```python
#Make one hot representation
from tensorflow.keras.preprocessing.text import one_hot
voc_size=5000
```

```python
onehot_rep=[one_hot(word,voc_size) for word in corpus]
onehot_rep
```

```python
print(corpus[3])
print(onehot_rep[3])
```

*Embeededding *

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#first we have to do padding
sent_maxlen=20
padding_data=pad_sequences(onehot_rep,padding='post',maxlen=sent_maxlen)
padding_data
```

```python
padding_data[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
#Create a model with embedding
vector_len=40
model=Sequential()
model.add(Embedding(voc_size,vector_len,input_length=sent_maxlen))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
len(padding_data),y.shape
```

```python
y=y[:8000]
```

```python
y.shape
```

```python
X_final=np.array(padding_data)
y_final=np.array(y)
```

```python
print(X_final.shape,y_final.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
